# The reparameterization trick

Intuitively we can think of reparameterization trick as follows:
Because we cannot compute the gradient of an expectation, we "move" the parameters of the probability distribution form the distribution space to the expectation space. In other words, we want to rewrite the expectation so that the distribution is independent of the parameter $\theta$. Then we simply take the gradient as we did for the model parameters.

This abstract idea can be formulated as transforming a sample from a fixed, known distribution to a sample form $q_\Phi(z)$. If we consider the Gaussian distribution, we can express z with respect to a fixed $\epsilon$, where $\epsilon$ follows the normal distribution: $z = \mu + \sigma \epsilon$ with $\epsilon \sim \mathcal{N}(0,1)$.
The $\epsilon$ term introduces the stochastic part and is not involved in the training process.
In a fully stochastic operation, you cannot perform backpropagation. So, instead, we keep a fixed part stochastic with $\epsilon$ and train the mean and the standard deviation.
